```python
import random
Category =['Book','Stationery','Office','Gift']
Products=[]

for i in range(random.randint(10,15)):
    id=i+1
    Products.append((id,"ITEM"+str(id),random.choice(Category),random.randint(50,300)))

for item in Products:
    print(item)
```

```python
MaxList=max(Products,key=lambda price:price[3])

for item in Products:
    if item[3]==MaxList[3]:
        print(item)
```

```python
MinList=min(Products,key=lambda price:price[3])

for item in Products:
    if item[3]==MinList[3]:
        print(item)
```

```python
for select in Category:
    count=0
    temp=[]
    for item in Products:
        if select == item[2]:
            count+=1
            temp.append(item)
    print(select,"=",count)
    for item in temp:
        print(item)
```

```python
x=input('จำนวนลูกค้า')
m=input('จำนวนรายการสูงสุด')
History=[]
for order in range(int(x)):
    print("Order",order)
    Order=[]
    for item in range(random.randint(1,int(m))):
        item=random.choice(Products)
        History.append(item)
        Order.append(item)
    Order.sort(key=lambda i:i[0])
    sum=0
    for item in Order:
        print(item)
        sum+=item[3]


    count={}
    price={}

    for id,name,cat,p in History:
        if id not in count:
            count[id]=0
            price[id]=p
        count[id]+=1
    # print(count)

    total = 0
    total_D=0
    for id in count:
        qty = count[id]        # จำนวน
        unit_price = price[id] # ราคาต่อชิ้น
        subtotal = qty * unit_price

        # ถ้าซื้อ 3 ชิ้นขึ้นไป ลด 15%
        if qty >= 3:
            discount = subtotal * 0.15
        else:
            discount = 0

        net = subtotal - discount
        total += net
        total_D += discount
    print("Sum =",sum)    
    print("Discount =",total_D)
    print("Net =",total)
    

```
