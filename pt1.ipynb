```python
import random

Cat =["Drink","Food","General","Cloth"]
Products=[]

for i in range(random.randint(10,15)):
    id=i+1
    Name=str("Product"+str(id))
    Category=random.choice(Cat)
    Price=random.randint(10,30)
    Products.append((id,Name,Category,Price))

for item in Products:
    print(item)
```

```python
print('Highest Price')
MaxpriceProduct=max(Products,key=lambda price:price[3])

for item in Products:
    # print(MaxpriceProduct[3])
    if item[3] == MaxpriceProduct[3]:
        print(item)
```

```python
print('Lowest Price')
MinpriceProduct=min(Products,key=lambda price:price[3])

for item in Products:
    # print(MaxpriceProduct[3])
    if item[3] == MinpriceProduct[3]:
        print(item)
```

```python
print('รายการสินค้าแต่ละประเภท')
for select in Cat:
    count=0
    temp=[]
    for item in Products:
        if select == item[2]:
            count+=1
            temp.append(item)
    print(select,":",count)
    for item in temp:
        print(item)
```

```python
x=input('จำนวนลูกค้า')
m=input('จำนวนสินค้า')
print('จำนวนลูกค้า :'+str(x))
print('จำนวนสินค้า :'+str(m))

History=[]
for customer  in range(int(x)):
    print("Order No."+str(customer))
    print("ID NAME CATE PRICE")
    temp=[]
    for item in range(random.randint(1,int(m))):
        x=random.choice(Products)
        temp.append(x)
        History.append(x)

    temp.sort(key=lambda id:id[0])
    for item in temp:
        print(item)

    # Sum
    SumProduct=0
    for item in temp:
        SumProduct+=item[3]
    print("Sum = "+str(SumProduct))

    # Discount
    tempcount = set()
    discount=0
    for item in temp:
        if item in tempcount:
            discount=(int(SumProduct)*10)/100
            break
        else:
            tempcount.add(item)
    print("Discount = "+str(discount))

    print("Net = "+str(SumProduct-discount))
    print("--------------------")
```

```python
History.sort(key=lambda id:id[0])

print('Bestsell')
print("ID NAME Cate Price Num")
temp2=[]
for select in Products:
    count=0
    temp=[]

    for item in History:
        if select[0] == item[0]:
            count+=1
            temp.append(item)
        if count > 0:
            product_summary = (select[0], select[1], select[2], select[3], count)
    temp2.append(product_summary)
# for i in temp2:
#     print(i)
Bestsell=max(temp2,key=lambda q:q[4])
for item in temp2:
    # print(MaxpriceProduct[3])
    if item[4] == Bestsell[4]:
        print(item)
```

```python
Lowsell=min(temp2,key=lambda q:q[4])
for item in temp2:
    # print(MaxpriceProduct[3])
    if item[4] == Lowsell[4]:
        print(item)
```
